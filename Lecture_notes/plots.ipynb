```python
# Plots.
import matplotlib.pyplot as plt

# Fancier plots.
import seaborn as sns
```

```python
# Numerical arrays.
import numpy as np
```

$ f(x) = x^2 -4x + 4 $

```python
# The above function.
def f(x):
  return x**2 - (4.0 * x) + 4.0
```

```python
# Array of x values.
x = np.linspace(-1.0, 5.0, 601)
```

```python
# Calculate the y values.
y = f(x)
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Interpolate.
ax.plot(x, y)
```

## Plots of Different Functions

***

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-4.0, 4.0, 801)

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2 + 1.0.
ax.plot(x, x**2 + 1.0, label='$x^2 + 1$')

# Plot x^2 - 1.0.
ax.plot(x, x**2 - 1.0, label='$x^2 - 1$')

# Plot x^2 - 3.0.
ax.plot(x, x**2 - 3.0, label='$x^2 - 3$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-12.0, 12.0, 801)

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2 + x.
ax.plot(x, x**2 + x, label='$x^2 + x$')

# Plot x^2 - x.
ax.plot(x, x**2 - x, label='$x^2 - x$')

# Plot x^2 - 4.0x.
ax.plot(x, x**2 - (4.0 * x), label='$x^2 - 4x$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(1.0, 10.0, 1001)

# Plot log (natural).
ax.plot(x, np.log(x), label='$\ln(x)$')

# Plot log base 2.
ax.plot(x, np.log2(x), label='$\log_2(x)$')

# Plot log base 10.
ax.plot(x, np.log10(x), label='$\log_{10}(x)$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-3.0, 3.0, 1001)

# Plot x.
ax.plot(x, x, label='$x$')

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^3.
ax.plot(x, x**3, label='$x^3$')

# Plot x^4.
ax.plot(x, x**4, label='$x^4$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(0.0, 50.0, 1001)

# Plot x.
ax.plot(x, x, label='$x$')

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2.
ax.plot(x, 2**x, label='$2^x$')

# Turn on legend.
ax.legend();
```
